## 3. Create a DataFrame from the following data and display only the names of people older than 28.
Name: ['Alice', 'Bob', 'Charlie'], Age: [25, 30, 35], City: ['Accra', 'Kumasi', 'Tamale']

```python
import pandas as pd
data = {
    'Name': ['Alice', 'Bob', 'Charlie'],
    'Age': [25, 30, 35],
    'City': ['Accra', 'Kumasi', 'Tamale']
}
df = pd.DataFrame(data)
df[df['Age'] > 28]['Name']
```

**Explanation:**
We create a DataFrame and filter rows where 'Age' > 28, then display only the 'Name' column.

## 6. Plot a line graph of x = [1,2,3,4,5] and y = [2,4,6,8,10] with a title 'Line Graph Example'.

```python
import matplotlib.pyplot as plt
x = [1,2,3,4,5]
y = [2,4,6,8,10]
plt.plot(x, y)
plt.title('Line Graph Example')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

**Explanation:**
We use matplotlib to plot x and y, set the title, and label the axes.

## 7. Create a bar chart showing the sales of three products: Product A: 10, Product B: 15, Product C: 7

```python
products = ['Product A', 'Product B', 'Product C']
sales = [10, 15, 7]
plt.bar(products, sales)
plt.title('Sales of Products')
plt.xlabel('Product')
plt.ylabel('Sales')
plt.show()
```

**Explanation:**
We use matplotlib's `bar` function to plot sales for each product.

## 8. Using Seaborn, create a scatter plot of x = [1, 2, 3, 4, 5] and y = [5, 7, 4, 8, 6].

```python
import seaborn as sns
sns.scatterplot(x=[1,2,3,4,5], y=[5,7,4,8,6])
plt.title('Seaborn Scatter Plot')
plt.show()
```

**Explanation:**
We use Seaborn's `scatterplot` to visualize the relationship between x and y.

## 9. Using Seaborn, create a boxplot for the data scores = [55, 70, 65, 80, 60, 75, 85].

```python
scores = [55, 70, 65, 80, 60, 75, 85]
sns.boxplot(scores)
plt.title('Seaborn Boxplot')
plt.show()
```

**Explanation:**

## 10. Create a histogram of data = [5,7,8,5,9,3,4,6,7,8,5,9] using Seaborn with 5 bins and a blue color.

```python
data = [5,7,8,5,9,3,4,6,7,8,5,9]
sns.histplot(data, bins=5, color='blue')
plt.title('Seaborn Histogram')
plt.show()
```

**Explanation:**
We use Seaborn's `histplot` to create a histogram with 5 bins and set the color to blue.
